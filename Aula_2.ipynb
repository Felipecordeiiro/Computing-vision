### Pré-processamento

#### Conversão de cores 

A conversão de cores é uma das funcionalidades mais úteis e comuns da biblioteca OpenCV em Python. É possível converter uma imagem de um espaço de cor para outro, por exemplo, de RGB para escala de cinza ou de RGB para HSV.

Para converter uma imagem de uma cor para outra, você pode usar a função cv2.cvtColor(), que converte uma imagem de um espaço de cor para outro. Esta função recebe dois parâmetros: a imagem a ser convertida e o espaço de cor para o qual você deseja convertê-la. Abaixo está um exemplo básico de como converter uma imagem em escala de cinza:

```python
import cv2

# Carrega a imagem em BGR
img = cv2.imread('image.jpg')

# Converte a imagem em escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Exibe a imagem original e a imagem em escala de cinza
cv2.imshow('imagem', img)
cv2.imshow('imagem', gray)

# Aguarda que o usuário feche a janela
cv2.waitKey(0)

# Libera os recursos
cv2.destroyAllWindows()
```

Neste exemplo, a imagem é carregada em BGR usando a função cv2.imread(). Em seguida, a função cv2.cvtColor() é usada para converter a imagem em escala de cinza. A constante cv2.COLOR_BGR2GRAY é usada para especificar a conversão para escala de cinza.

Para converter uma imagem de RGB para HSV, você pode usar a seguinte linha de código:

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

Neste exemplo, a constante cv2.COLOR_BGR2HSV é usada para especificar a conversão para o espaço de cor HSV.

#### Redimensionamento de imagens
Redimensionar uma imagem é uma operação muito comum no processamento de imagens, e a biblioteca OpenCV em Python oferece várias funções para realizar essa operação.

A maneira mais simples de redimensionar uma imagem em OpenCV é usar a função cv2.resize(). Esta função recebe três parâmetros: a imagem que você deseja redimensionar, o tamanho desejado da imagem (largura e altura), e um parâmetro opcional para especificar o método de interpolação usado para redimensionar a imagem.

Abaixo está um exemplo de como redimensionar uma imagem em OpenCV em Python:

```python
import cv2

# Carrega a imagem
img = cv2.imread('image.jpg')

# Redimensiona a imagem para 640x480 pixels
resized_img = cv2.resize(img, (200, 200))

# Exibe a imagem original e a imagem redimensionada
cv2.imshow('', img)
cv2.imshow('', resized_img)

# Aguarda que o usuário feche a janela
cv2.waitKey(0)

# Libera os recursos
cv2.destroyAllWindows()
```

### Segmentação de imagens

A segmentação de imagens é uma tarefa fundamental em Visão Computacional que consiste em dividir uma imagem em regiões ou segmentos com características visuais similares. Essa técnica é amplamente utilizada em diversas áreas, como reconhecimento de objetos, análise de texturas, detecção de bordas, entre outras.

***Thresholding global***

O thresholding global é uma técnica de segmentação de imagens que define um valor de limiar global para a imagem inteira. Todos os pixels com valores acima do limiar são considerados como pertencentes ao objeto, e os pixels com valores abaixo do limiar são considerados como pertencentes ao fundo. 

Neste exemplo, a função cv2.threshold() é utilizada para aplicar o thresholding global na imagem. O valor 127 é utilizado como limiar, e o tipo de limiarização cv2.THRESH_BINARY é utilizado para segmentar a imagem em preto e branco. O valor retornado pela função cv2.threshold() é armazenado na variável ret, mas não é utilizado neste exemplo.

```python
import cv2

# Lê a imagem em escala de cinza
img = cv2.imread('image.jpg', 0)

# Aplica o thresholding global
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Exibe a imagem segmentada
cv2.imshow('',img)
cv2.imshow('',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

***Thresholding local (adaptativo)***

O thresholding local (adaptativo) é uma técnica de segmentação de imagens que define um valor de limiar diferente para cada região da imagem. Essa técnica é útil para segmentar imagens com variações de iluminação ou com regiões de intensidades muito distintas.

Neste exemplo, a função cv2.adaptiveThreshold() é utilizada para aplicar o thresholding local (adaptativo) na imagem. O parâmetro 255 é utilizado como limiar máximo, e o tipo de limiarização cv2.THRESH_BINARY é utilizado para segmentar a imagem em preto e branco. O parâmetro cv2.

ADAPTIVE_THRESH_GAUSSIAN_C indica que o limiar é calculado com base em uma média ponderada dos pixels vizinhos, com pesos definidos por uma função gaussiana. Os parâmetros 11 e 2 definem o tamanho da janela de vizinhança e a constante de subtração, respectivamente.

```python
import cv2

# Lê a imagem em escala de cinza
img = cv2.imread('image.jpg', 0)

# Aplica o thresholding local (adaptativo)
thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Exibe a imagem segmentada
cv2.imshow('',img)
cv2.imshow('',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

***Segmentação com o Método de Otsu***

O método de Otsu é uma técnica de segmentação de imagens que tenta encontrar o melhor valor de limiar possível para separar o objeto do fundo, considerando a distribuição dos níveis de cinza da imagem.

```python
import cv2

# Lê a imagem em escala de cinza
img = cv2.imread('image.jpg', 0)

# Aplica o método de Otsu
ret, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Exibe a imagem segmentada
cv2.imshow('',img)
cv2.imshow('',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Lê a imagem em escala de cinza
img = cv2.imread('pes.png', 0)
resize_img = cv2.resize(img,(750,720))

# Aplica o método de Otsu
ret, thresh = cv2.threshold(resize_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Exibe a imagem segmentada
cv2.imshow('',resize_img)
cv2.imshow('',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Extração de características

A extração de características é uma das principais etapas do processamento de imagem em visão computacional. A extração de características envolve a identificação e extração de informações relevantes e distintivas em uma imagem, que podem ser usadas para fins como reconhecimento de objetos, detecção de padrões, segmentação de imagem, entre outros.

#### Descritor de forma

Os momentos de Hu são um conjunto de sete números reais que são usados como descritores de forma em imagens.

Neste exemplo, a imagem é carregada em escala de cinza usando a função cv2.imread() do OpenCV com a flag cv2.IMREAD_GRAYSCALE. Em seguida, os momentos de Hu são calculados usando a função cv2.HuMoments(), que recebe como entrada os momentos da imagem calculados pela função cv2.moments(). Os momentos de Hu são normalizados para ter somente valores positivos, de acordo com a fórmula matemática de cálculo desses momentos. Por fim, os momentos de Hu normalizados são impressos na saída do terminal.

```python
import cv2
import numpy as np

# Carrega a imagem em escala de cinza
img_gray = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Calcula os momentos de Hu
hu_moments = cv2.HuMoments(cv2.moments(img_gray)).flatten()

# Imprime os momentos de Hu
print(hu_moments)
```

***Normalizado***

```python
import cv2
import numpy as np

# Carrega a imagem em escala de cinza
img_gray = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Calcula os momentos de Hu
hu_moments = cv2.HuMoments(cv2.moments(img_gray)).flatten()

# Normaliza os momentos de Hu para ter somente valores positivos
for i in range(0,7):
    hu_moments[i] = abs(hu_moments[i])

# Imprime os momentos de Hu
print(hu_moments)
```

#### Extração de recursos

***LBP***

LBP (Local Binary Pattern): O LBP é um descritor de textura que codifica a relação entre um pixel e seus vizinhos em uma imagem.


Neste exemplo, a imagem é carregada em formato RGB e convertida para escala de cinza antes de calcular o descritor LBP usando a função local_binary_pattern. São definidos os parâmetros de raio, número de pontos e método de codificação para o descritor LBP. Em seguida, o histograma normalizado é calculado a partir do descritor LBP e impresso na saída do terminal.

```python
import numpy as np
import skimage.color as skcolor
import skimage.feature as skfeat
import skimage.io as io

# Carrega a imagem em escala de cinza
img_gray = skcolor.rgb2gray(io.imread('image.jpg'))

# Converte a imagem para o tipo de dado adequado
img_gray = np.uint8(img_gray * 255)

# Define os parâmetros para o descritor LBP
radius = 3
n_points = 8 * radius
method = 'uniform'

# Calcula o descritor LBP usando a função local_binary_pattern
lbp = skfeat.local_binary_pattern(img_gray, n_points, radius, method)

# Obtém o histograma de ocorrências dos padrões LBP
(hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))

# Normaliza o histograma para ter somente valores entre 0 e 1
hist = hist.astype('float')
hist /= (hist.sum() + 1e-7)

# Imprime o histograma normalizado
print(hist)
```
